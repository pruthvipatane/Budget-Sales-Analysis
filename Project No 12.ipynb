```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('C:/Users/PRUTHVI/Desktop/COLLEGE/University/Project Work/UnifiedMentor/Budget Sales data/Budget.xlsx')
```

```python
print(df)
```

```python
summary = df.describe()
print(summary)
```

```python
# Plotting total sales by month
monthly_sales = df[df.columns[3:-1]].sum()
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales)
plt.title('Monthly Sales Trend')
plt.ylabel('Sales ($)')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()
```

```python
# Total sales by category
total_sales_by_category = df.groupby('Category')['Grand Total'].sum()

# Total sales by subcategory
total_sales_by_subcategory = df.groupby('Subcategory')['Grand Total'].sum()
```
